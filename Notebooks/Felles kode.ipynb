```python
# Nødvendige importer:
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""Konstanter og parametre"""

# kappe/usynlig person:
h1 = 1.5 # [m] høyde under kappe
sigma1 = 1 # [m] std. innerste kappe
x1 = 4 # [m] 1/2 bredde på innerste kappe
N = 12 # antall lag 

# (ikke-)observant:
H = 1.7 # [m] høyde på Voldemort
L = 40 # [m] avstand fra Harry
L_ = L - x1 # avstand fra kappe
```

```python
# Analytisk - innerste lag, n=1

def GaussKurve(x, sigma, A):
    y = A * np.exp(-x**2 / (2*sigma**2))
    #dx = x[1]- x[0]
    #yder = np.gradient(y, dx)
    yder = -x/sigma**2 * y
    
    return x, y, yder



def BrytningsIndeks(x, yder, theta):
    # theta: innfallsvinkel 0 < theta < pi/2 fra luft til kappe
    K = -np.sin(theta)
    n = K * np.sqrt(1 + yder**2) / yder
    n[np.abs(x) <0.01] = float('nan')
    return x, n, K

```

```python
x = np.linspace(-x1, x1, 1000)
#x, y, yder = GaussKurve(x, sigma1, h1)

y = -h1*x**2 + h1
yder = np.gradient(y, x[1]-x[0])
x, n, K = BrytningsIndeks(x, yder, np.pi/3)
print(K)
plt.plot(x, y)
plt.plot(x, yder)
plt.axis([-1, 0, 0, 2])
plt.show()

plt.plot(x, n)
plt.axis([-x1,0, -1.2, 1.2])
plt.show()
plt.plot(x, n)
plt.show()
```

```python
x = np.linspace(-x1, x1, 1000)

y = np.cos(x * np.pi/(x1*4/5) ) * h1/2 + h1/2
yder = np.gradient(y, x[1]-x[0])
x, n, K = BrytningsIndeks(x, yder, np.pi/3)
print(K)
plt.plot(x, y)
plt.plot(x, yder, ls='--')
plt.grid()
plt.show()

plt.plot(x, n)
plt.axis([-x1, 0, -2.5, 2.5])
plt.show()
plt.plot(x, n)
plt.show()
```
